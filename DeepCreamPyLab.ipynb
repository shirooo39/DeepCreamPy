```
# ============================= FORM ============================= #
# @markdown <p><font size=5px>← [Start] DeepCreamPy</font></p></br>
image_source = 'upload' #@param ["upload", "url", "vm"]
# @markdown > upload = Upload image from your local disk.</br>url = Get image from the internet (type in direct url to the image).</br>vm = Image already stored in DeepCreamPy's input folder in the VM's disk.</br></br><font color=red>The file name of both of the green-colored and original images have to be the same!</font></br></br><font size=2px>(simply choose "vm" as the image source to ignore the image_source form)
censor_type = 'bar' #@param ["bar", "mosaic"]
automatically_clear_input_folder = False # @param {type:"boolean"}
# ================================================================ #

from IPython.display import clear_output
import google.colab.files

txt1 = 'Upload an image file which the censored part(s) has been colored green (0, 255, 0).\n'
txt2 = 'Upload the original censored image.\n'
txt3 = 'Enter the direct URL of the image which the censored part(s) has been colored green (0, 255, 0).\nType exit to cancel.\n'
txt4 = 'Enter the direct URL of the original censored image.\nType exit to cancel.\n'
txt5 = 'File upload has been canceled!'
txt6 = 'DeepCreamPy has finished its job!\nAll of the processed images are stored in /content/DeepCreamPy/decensor_output'

if censor_type == 'bar':
  if image_source == 'upload':
    %cd /content/DeepCreamPy/decensor_input
    clear_output()

    print(txt1)
    uploaded = google.colab.files.upload()

    if not uploaded:
      clear_output()
      print(txt5)
    else:
      %cd /content/DeepCreamPy
      clear_output()

      !python decensor.py

      clear_output()
      print(txt6)
  elif image_source == 'url':
    print(txt3)
    url = input("")

    if url == 'exit' or url == 'Exit' or url == 'EXIT':
      clear_output()
      exit
    else:
      %cd /content/DeepCreamPy/decensor_input
      clear_output()

      !wget {url}
      clear_output()

      %cd /content/DeepCreamPy
      clear_output()

      !python decensor.py

      clear_output()
      print(txt6)
  elif image_source == 'vm':
    %cd /content/DeepCreamPy
    clear_output()

    !python decensor.py

    clear_output()
    print(txt6)
elif censor_type == 'mosaic':
  if image_source == 'upload':
    %cd /content/DeepCreamPy/decensor_input
    clear_output()

    print(txt1)
    uploaded = google.colab.files.upload()
    clear_output()

    if not uploaded:
      print(txt5)
    else:
      %cd /content/DeepCreamPy/decensor_input_original
      clear_output()

      print(txt2)
      uploaded = google.colab.files.upload()

      if not uploaded:
        print(txt5)
      else:
        %cd /content/DeepCreamPy
        clear_output()

        !python decensor.py --is_mosaic 1

        clear_output()
        print(txt6)
  elif image_source == 'url':
    print(txt3)
    url1 = input("")

    if url1 == 'exit' or url1 == 'Exit' or url1 == 'EXIT':
      clear_output()
      exit
    else:
      %cd /content/DeepCreamPy/decensor_input
      clear_output()

      !wget {url1}
      clear_output()

      print(txt4)
      url2 = input("")

      if url2 == 'exit' or url2 == 'Exit' or url2 == 'EXIT':
        exit
      else:
        %cd /content/DeepCreamPy/decensor_input_original
        clear_output()

        !wget {url}
        clear_output()

        %cd /content/DeepCreamPy
        clear_output()

        !python decensor.py --is_mosaic 1

        clear_output()
        print(txt6)
  elif image_source == 'vm':
    %cd /content/DeepCreamPy
    clear_output()

    !python decensor.py --is_mosaic 1

    clear_output()
    print(txt6)

if automatically_clear_input_folder == True:
  %rm -rf /content/DeepCreamPy/decensor_input/*
  %rm -rf /content/DeepCreamPy/decensor_input_original/*
```
